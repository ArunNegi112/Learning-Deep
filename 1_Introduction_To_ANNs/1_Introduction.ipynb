ANN is a Machine Learning model inspired by network of neurons in human brain. This was the initial idea of ANNs, but over the time, ANNs have became logically very different from neurons in our brains (Though we still use the same term)

### Artificial Neuron
__Warren McCulloh and Walter Pitts__ proposed a simple model of biological neuron that works on the basis of _Proposition log_ and is later known as ___artificial neuron___.

This neuron has one or more binary inputs (on/off) and one binary output. Even though its such a simple model it is possible to build a complex network of neurons that can solve any proposition logic.

Example: 

1. C activates when A is activated
2. C activates when A and B both are activated
3. C activates when either A or B is activated
4. C activates when only A is activated 

___These networks can be combined to compute complex logical problem___

### The Perceptron 
- One of the simplest ANN structure
- It is based on an artificial neuron, called _"TLU (Threshold Logic Unit) or LTU (Linear Threshold Unit)"_ 
- How TLU works:
    - TLU have input and output as numbers (not binary values)
    - Each input connection is associated with a _weight_. 
    - The we compute its weighted sum: $z = x^Tw = x_1w_1 + x_2w_2 + ...$
    - Then we use _step function_ on the weighted sum and get the output: $h_w(x) = step(z)$


Note: _The _artificial neuron_ here is a single unit _in_ Perceptron. Perceptron is basically a Supervised Machine Learning Model for classification, Which is similar to biological neuron in our brain and became the core block of Deep Learning_  
Shown below is a artificial neuron based on TLU

Most common step function is: _Heaviside_   
And sometimes _sgn_ function is used

### Structure of Single layer Perceptron

Example of a single layer perceptron with 2 input neurons and 3 outputs neurons, This perceptron can classify inputs in 3 different binary classes (means class A yes/no, class B yes/no, class C yes/no), hence this is a multioutput classifier.

- A Perceptron is basically a single layer of TLUs. Where each TLU is connected to each input. 
- Inputs of a Perceptron are fed to a layer called _Input layer_. Input layer consists of inputs and a special neuron called __"Bias neuron"__ (which outputs 1 always).
- This input layer takes inputs, and a bias neuron and give them to output layer without any change.
- If all neurons in a layer are connected to all neurons in its previous layer (here input layer) then it is called a __Fully connected layer__ or __Dense layer__ 

___Computing the outputs___:    
$$h_{w,b}(X) = \phi(XW + b)$$


Where: 
- $X$ is input matrix. 
    - Shape(X) = $m \times n$. where, m= number of instances (inputs), n= number of features= number of input neurons
- $W$ is matrix of connection weights of inputs.
    - Shape(W) = $n \times n_{neuron}$. Where, n= number of features= number of input neurons, n_neurons= number of output neurons
- $XW$ is weighted sum of inputs. 
    - Shape(XW) = $m \times n_{neurons}$
- $b$ is bias vector. It contains weights between bias neuron and artificial neuron.
    - Shape(b) = $1 \times n_{neurons}$. To be added with $XB$ it will broadcast into $m \times n_{neurons}$

- Function $\phi$ is _Activation function_. If artificial neurons are TLUs, it is _Step function_. 


## Demo

```python
#Load the data (ignore)

from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
print(iris.keys())

#I will try binary classification for simplicity here: setosa, versicolor 
#And two features: petal_length and petal_width
print("Original targets:", iris['target'].unique())
print("Original targets:", iris['target_names'])
print("Features: ", iris['data'].columns)
data = iris['frame'][iris['frame']['target']!=2][['petal length (cm)','petal width (cm)', 'target']]
print("We using: ", data['target'].unique())
print("We using: ", "['setosa' 'versicolor']")
```

```python
data
```

```python
X = data.iloc[:,0:2].values
y = data.iloc[:,-1].values
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='coolwarm')
```

```python
from sklearn.linear_model import Perceptron

perceptron = Perceptron()
perceptron.fit(X,y)
```

```python
print("Weights: " ,perceptron.coef_)
print("Bias: ", perceptron.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
import numpy as np
plot_decision_regions(np.array(X),np.array(y),clf=perceptron)
```
