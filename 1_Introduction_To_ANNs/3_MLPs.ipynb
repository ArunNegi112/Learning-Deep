### What is a Neural Network Architechture?
Neural Network Architechture is arrangement of connection of multiple artificial neurons organized in layers working together to process data and get desired output

## Multi-Layer Perceptron 
- A neural network architechture made of multiple layers of perceptrons(neurons).
- A MLP includes: An input layer, One or more hidden layers, An output layer  
- Each layer except output layer includes a bias neuron  
- Each neuron(except bias neuron) in a layer receives input from all neurons in previous layer 
- And each neuron(including bias neuron) in a layer is fully connected to next layer.

### _How MLP architechture solves the limitations of Perceptron?_

___The abstract idea behind MLP is___   
- Each perceptron (neuron) will find a set of parameters (decision boundary graphically).
- If we take a linear combination of those parameters from the previous layer neurons, compute their weighted sum, and pass it as input to the next layer (_kinda_ superimposing decision boundaries graphically) 
- The next layer of neurons will combine these simpler decision boundaries to form more complex and nonlinear decision boundaries.

### __Notation__
In training of MLP, we have to find following quantities: _weights_ for each node of each layer, _bias_ for each layer, _output_ from each node in each layer
- $X_{m \times n}$ : Input data with $m$ instances and $n$ features
- $b_{ij}$ : bias in $i_{th}$ layer and $k_{th}$ node
- $O_{ij}$ : output of $i_{th}$ layer and $j_{th}$ node
- $w^k_{ij}$ : weight in $k_{th}$ layer, coming from $i_{th}$ node, and going to $j_{th}$ node

Also, lets check how many parameters we have to find in an architecht:   
Take the example from above "MLP of one hidden layer and four nodes", there we have to find: 
$$
 ((4+4) + 4) + ((1+1+1+1)+1) = 17 \text{   parameters}
$$
8 weights + 4 bias from input layer to hidden layer, Then, 4 weights + 1 bias from hidden layer to output layer

## _Forward Propagation_
Lets take an example of the process:

Consider the architecture:  


- $X_{m \times 4}$ is input data
- we using sigmoid function internally as activation function: $\sigma(X^T.W+ b)$
- Input is a single instance matrix $X_{4 \times 1}$
- from input to hidden layer: 
    - output matrix = $\sigma(X^T_{(1\times 4)}.W_{(4\times 3)} + b_{(1\times 3)})$ ---> $\sigma (1\times 3)$ ---> $O_{1 \times 3}$ output matrix

- from 2nd to output layer
    - output matrix = $\sigma (O_{1 \times 3}. W_{3 \times 2} + b_{1 \times 2})$  ---> $\sigma (1\times 2)$ ---> $O_{1\times 2}$ output matrix

- Final output
    - final output = $\sigma (O_{1 \times 2}. W{2\times 1} + b_{1\times 1})$ ---> $\sigma (1 \times 1)$ ---> $O_{1\times 1}$
